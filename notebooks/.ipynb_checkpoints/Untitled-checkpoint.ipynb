```python
import json
import os 
import random

import matplotlib.pyplot as plt
import matplotlib.patches as patches

from glob import glob
from pathlib import Path
from tqdm import tqdm
from PIL import Image
```

```python
with open("../classes.txt") as f:
    classes = eval(f.read())

print(f"{classes=}")
```

```python
train = Path("../train")
test = Path("../validation")

train_ann = train / "annos"
test_ann = test / "annos"

img = None

anns = list(test_ann.glob("*"))
random.shuffle(anns)

for ann in anns:
    
    fname = ann.stem
    img = (test / "image" / fname).with_suffix(".jpg")
    
    with open(ann) as f:
        data = json.load(f)
    break

n = 1
boxes = []

while f"item{n}" in data:
    item = data[f"item{n}"]
    boxes.append({
        "cords": item["bounding_box"],
        "class": item["category_name"]
    })
    n += 1
    
img = Image.open(img)

fig, ax = plt.subplots()
ax.imshow(img)
for box in boxes:
    cords = box["cords"]
    rect = patches.Rectangle((cords[0], cords[1]), cords[2] - cords[0], cords[3] - cords[1], linewidth=1, facecolor='none', edgecolor='r')
    ax.add_patch(rect)
plt.show()
    
```

```python
(train / "yolo_annos").mkdir(parents=True, exist_ok=True)
(test / "yolo_annos").mkdir(parents=True, exist_ok=True)

for ann in tqdm(train_ann.glob("*")):
    fname = ann.stem
    img = (train / "image" / fname).with_suffix(".jpg")
    img = Image.open(img)
    width, height = img.size
    
    with open(ann) as f:
        data = json.load(f)
    
    n = 1
    with open((train / "yolo_annos" / ann.stem).with_suffix(".txt"), "a+") as f1:
        while f"item{n}" in data:
            item = data[f"item{n}"]
            x1, y1, x2, y2 = item["bounding_box"]
            class_index = classes.index(item["category_name"])
            xc = ((x1 + x2) // 2) / width
            yc = ((y1 + y2) // 2) / height
            w = (x2 - x1) / width
            h = (y2 - y1) / height
            f1.write(f"{class_index} {xc} {yc} {w} {h}\n")
            
        
    
            n += 1
    
    break
    
    
    
        
    
```
